## Web Scrapping Project 

```python
import requests
from bs4 import BeautifulSoup
import time

##Getting Brand details
def brand_scrap(link_url):
    c = 1
    brand_code=requests.get(link_url,headers = {'User-agent': 'Super Bot 9000'})
    brand_plain_text=brand_code.text
    brand_soup = BeautifulSoup(brand_plain_text)
    for phone in brand_soup.find_all('div',{'class':'brandmenu-v2 light l-box clearfix'}):
        for li in phone.find_all('li'):
            for anc in li.find_all('a'):
                anc_src = r'http://www.gsmarena.com/' + anc.get('href')
                brand_name = anc.string
                print(c, brand_name,"\n", anc_src, "\n")
                c += 1
                phone_scrap(brand_name, anc_src)

##Getting Phone Details
def phone_scrap(name, hrefs):
    time.sleep(1)
    model_code=requests.get(hrefs,headers = {'User-agent': 'Super Bot 9000'})
    model_plain_text=model_code.text
    model_soup = BeautifulSoup(model_plain_text)
    for model in model_soup.find_all('div',{'class':'makers'}):
        for li in model.find_all('li')[:10]:
            for anc in li.find_all('a'):
                anc_src = r'http://www.gsmarena.com/' + anc.get('href')
                for nam in (sp.find('span') for sp in anc.find_all('strong')):
                    model_name = nam.string
                    print(model_name)
                    model_scrap(name,model_name,anc_src)

##Gettting Model Specifications
def model_scrap(brand_name,model_name,link_url):
    spec_list = []
    time.sleep(1)
    source_code=requests.get(link_url,headers = {'User-agent':'Super Bot 9000'})
    plain_text=source_code.text
    soup = BeautifulSoup(plain_text)
    soup2=BeautifulSoup(str(soup.select('#specs-list > table')))
    if (soup2.find('td',{'data-spec':'os'})):
        temp_os = soup2.find('td',{'data-spec':'os'})
        os = temp_os.text
    else:
        os = 'NA'
    if (soup2.find('td',{'data-spec':'gpu'})):
        temp_gpu = soup2.find('td',{'data-spec':'gpu'})
        gpu = temp_gpu.text
    else:
        gpu = 'NA'
    if (soup2.find('td',{'data-spec':'cpu'})):
        temp_cpu = soup2.find('td',{'data-spec':'cpu'})
        cpu = temp_cpu.text
    else:
        cpu = 'NA'
    if (soup2.find('td',{'data-spec':'internalmemory'})):
        temp_internal_memory = soup2.find('td',{'data-spec':'internalmemory'})
        internal_memory = temp_internal_memory.text
    else:
        internal_memory = "NA"
    if (soup2.find('td',{'data-spec':'cam1modules'})):
        temp_main_camera = soup2.find('td',{'data-spec':'cam1modules'})
        main_camera = temp_main_camera.text
    else:
        main_camera = "NA"
    if (soup2.find('td',{'data-spec':'cam2modules'})):
        temp_selfie_camera = soup2.find('td',{'data-spec':'cam2modules'})
        selfie_camera = temp_selfie_camera.text
    else:
        selfie_camera = "NA"
    if (soup2.find('td',{'data-spec':'sensors'})):
        temp_sensors = soup2.find('td',{'data-spec':'sensors'})
        sensors = temp_sensors.text
    else:
        sensors = "NA"
    if (soup2.find('td',{'data-spec':'batdescription1'})):
        temp_battery = soup2.find('td',{'data-spec':'batdescription1'})
        battery = temp_battery.text
    else:
        battery = "NA"
    if (soup2.find('td',{'data-spec':'price'})):
        temp_price = soup2.find('td',{'data-spec':'price'})
        price = temp_price.text
    else:
        price = "NA"
        
    spec_list = [brand_name,model_name, os, main_camera, selfie_camera, 
                 gpu, cpu, internal_memory, sensors, battery, price]
    main_list.append(spec_list)

    
    
main_list = [['BRAND','MODEL','OS','PRIMARY CAMERA','SELFIE CAMERA', 'GPU', 'CPU', 'INTERNAL MEMORY', 'SENSORS', 'BATTERY', 'PRICE']]
brand_scrap(r'http://www.gsmarena.com')
print(main_list)
```

```python
#Writing into data set.
import csv
with open("Dataset.csv", "w", newline="", encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(main_list)
```
